## Problem description <a name="introduction"></a>

Five years ago - in 2015 - after our studies my wife and me decided to **move to Munich**.
After a couple of weeks, we both found a great job in the city and signed the respective contracts.
So far so good. The questions was: **in which area of the city should we look for a flat?**

I think this a very common problem which many people are facing. Especially when you are new to the city and you have no idea about how the different areas look like, how you like the people there and how attractive it is to for activities after work in your leisure time. In most of the cases you will google a little bit or ask some colleagues and friends. But at the end there will be so many options, so many different opinions - at the end you are just as smart as before and you will take some gut feeling decision. At least for us it was like this back then.

Maybe you had a similar problem some time in the past? Probably it will occur in the future?

In this project I will try to find the **optimal location in Munich** (Germany) for our situation back then.
Keeping in mind, that the company of my wife is north of Munich and my job is located in the south. As we both do not want spend much time for travelling to work on a daily basis the distance to should be kept short. In addition rental prices and the density of restaurants and cafes plays also a big role.

I will use the data science skills I learned so far in this course to generate the most promissing neighborhoods based on this criteria. In addition this should be supported by a geo visualization of Munich.

## Data <a name="data"></a>

After describing the problem in the first place, the queston is: which data is required and relevant to support the decision for the right neighborhood in Munich?

When I talked to my stakeholder - my wife - we identified the following factors that would have influenced our decision in 2015:
* a) distance of neighborhood from company of my wife
* b) distance of neighborhood from my company
* c) "attractivity" of the neighboorhood (Restaurants? Bars? Activities?)
* d) rental price



The first two factors - distance of neighborhood to companies (a + b) -  will be calculated with help of the geopy library. I will measure the distances from the centre of the neighborhood with the Vincenty’s formulae, which is a more exact way to calculate distances on earth since it takes into account that the earth is an oblate spheroid. The lower the sum of both distances from the neighborhood to our job, the better. Both distances will be weighted equally.

Unfortunately unlike for New York or Toronto, there is not much data available for the city of Munich which can be used to measure the attractivity of a neighborhood (c) and solve this problem. 
The data source which contains most of datasets regarding munich is **opengov-muenchen.de** (link https://www.opengov-muenchen.de/tl/dataset?page=2).
On this website there are around 150 datasets. However most of them are in general not suitable for this project (e.g. statistics regarding „Oktoberfest“ or building permits). Others might be interesting („theatre“, „leisure time“, „cinema“ data) but have limited information. After scanning all of the datasets, I identified 3 of them as potentially relevant for people who want to relocate:
* Average age of population
* Share of immigrants
* Unemployment rate


Maybe for other  people it could be a decisive factor, but for us the share of immigrants does not matter at all.  Also the number of unemployed people of the neighborhood plays only a minor role. However, the third variable - **average age of population** - could be interesting, as to us it is also correlated with attractivity of the neighborhood. In most of the cases, the younger the average age, the more students and young adults are living in the area. On the one hand this could have a positive impact on the number of events, bars etc. On the other hand as we were both in our late 20's, the potential neighbors interests are more likely to match with ours. (noise level..)


In addition to that I will also use **Foursquare location data** to measure the attractivity of the neighborhood and solve the described problem. 
The higher the number of restaurants, the more attractive the area is for us. The number of restaurents and their type and location in every neighborhood will be obtained using Foursquare API.


Last but not least, the probably most decisive factor is the **average rental price**. This will be obtaned by using data from "Suedbayrische Immobiliengesellschaft": https://suedbayerische-immobilien.de/Mietpreise-Muenchen-Stadtteile

From the first glance at the data, I assume that Data Cleaning will probably not play a big role in this project. For example, there are no missing values in the data sources described before. Therefore it will be handled in the methodology part of the project.
